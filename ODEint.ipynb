## Part 1: Modeling a system using ODEINT


Suppose that fluid drains from tank A to B at rate r,  drains from tank B to C at rate r, then drains from tank C to A at rate r. The tank volumes remain constant due to constant recycling of fluid. Uniform stirring of each tank is assumed, which implies uniform salt concentration throughout each tank.

Let $x_1(t)$, $x_2(t)$, $x_3(t)$ denote the amount of salt at time $t$ in each tank. No salt is lost from the system, due to recycling. Using compartment analysis, the recycled cascade is modeled by the non-triangular system

\begin{align}
\frac{dx_1}{dt}  & = −\frac{1}{6}x_1 + \frac{1}{6}x_3  \\
\frac{dx_2}{dt}  & = \frac{1}{6}x_1 - \frac{1}{3}x_2 \\
\frac{dx_3}{dt}  & = \frac{1}{3}x_2 - \frac{1}{6}x_3\\
\end{align}

* (24 points) a. Using ODEINT, solve the system using the initial condition of $x_1(0)=10$, $x_2(0)=0$, $x_3(0)=0$. Start with $t=0$ and end with $t=30$ using increments of $\Delta t = 0.1$. 


* (5 points) b. After you solve the system, make one plot of the amount of salt in each pond as a function of time. Include a legend in the plot.


* (5 points) c. Which terms in your ODEs would you need to change to slow the rate at which salt is transferred from tank B to tank C?

```python
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

def saltfunc(y,t):
    x1 = y[0]
    x2 = y[1]
    x3 = y[2]
    
    dx1 = - (1/6)*x1 + (1/3)*x3
    dx2 = (1/6)*x1 - (1/3)*x2
    dx3 =  (1/3)*x2 - (1/6)*x3
    return[dx1,dx2,dx3]

# initial conditions  and time
x1_init = 10
x2_init = 0
x3_init = 0 
t = np.linspace(0,30,301)
y0 = [x1_init,x2_init, x3_init]

# solving
sol = odeint(saltfunc, y0, t)

X1 = sol[:,0]
X2 = sol[:,1]
X3 = sol[:,2]
```

```python
# Put your answer to Part 1b here
plt.plot(t,X1,label = "Tank A" )
plt.plot(t, X2,label = "Tank B")
plt.plot(t,X3,label = "Tank C ")
plt.title(" Amount of salt in each Tank vs Time")
plt.xlabel("Time")
plt.ylabel("Amount of salt")
plt.grid()
plt.legend()
plt.show()
```
